```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10
m = 101

x = np.linspace(0.0, n, m)
y = x[::-1]

u = []
v = []

u.append(0.0)
v.append(y[0])

for i in range(m - 2):
    x_1_1, x_2_1 = 0.0, x[i + 1]
    y_1_1, y_2_1 = y[i], 0.0
    m_1 = (y_2_1 - y_1_1) / (x_2_1 - x_1_1)
    
    x_1_2, x_2_2 = 0.0, x[i + 2]
    y_1_2, y_2_2 = y[i + 1], 0.0
    m_2 = (y_2_2 - y_1_2) / (x_2_2 - x_1_2)
    
    x_p = ((m_1 * x_1_1 - y_1_1) - (m_2 * x_1_2 - y_1_2)) / (m_1 - m_2)
    y_p = m_1 * (x_p - x_1_1) + y_1_1
    
    u.append(x_p)
    v.append(y_p)
    
u.append(x[-1])
v.append(0.0)
```

```python
t = np.linspace(0.0, n, m)
s = - n * np.sqrt(1.0 - (((t / n) - 1.0) ** 2.0)) + n
```

```python
alpha = 10.12
p = np.linspace(0.0, n, m)
q = alpha * ((1.0 - n / alpha) ** (p / n) - 1.0) + n
```

```python
plt.figure(figsize = (10, 8))
plt.plot(u, v, "-", linewidth = 2, label = "star")
plt.plot(t, s, "-", linewidth = 2, label = "circle")
plt.plot(p, q, "-", linewidth = 2, label = "exponential")
plt.grid()
plt.legend()
plt.axis("equal")
```
